````python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
````

````python
Here's how you might implement this using Python's pandas library to handle the data and matplotlib to plot the results. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# assuming the data is in a pandas DataFrame 'df'
df = pd.DataFrame({
    'fips': ['01', '02', '04', '05', '06', '05', '06', '08', '10', '11'],
    'name': ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Arkansas', 'California', 'Colorado', 'Delaware', 'District of Columbia'],
    'income': [36879, 56234, 44282, 34999, 53629, 48952, 80440, 77127, 70176, 92266],
    'year': [2005, 2005, 2005, 2005, 2005, 2019, 2019, 2019, 2019, 2019]
})

# calculate the income growth rate for each state
df_2005 = df[df['year'] == 2005].set_index('name')
df_2019 = df[df['year'] == 2019].set_index('name')
growth_rate = (df_2019['income'] - df_2005['income']) / df_2005['income']

# plot the results
growth_rate.sort_values().plot(kind='barh', figsize=(10, 6))
plt.title('Income Growth Rate from 2005 to 2019')
plt.xlabel('Growth Rate')
plt.ylabel('State')
plt.grid(axis='x')
plt.show()
```

Please replace the DataFrame 'df' with your actual DataFrame. This script calculates the growth rate as `(income_2019 - income_2005) / income_2005` for each state, then plots a horizontal bar chart sorted by the growth rate.
````

````python
import matplotlib.pyplot as plt

# Reorder the dataframe based on state and year
df = df.sort_values(["name", "year"])

# Group by state and calculate the income growth rate
df["income_growth_rate"] = df.groupby("name")["income"].pct_change() * 100

# Pivot the dataframe to wide format for plotting
df_pivot = df.pivot(index='year', columns='name', values='income_growth_rate')

# Plot the data
plt.figure(figsize=(12,8))
for column in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Income growth rate by state", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Income growth rate (%)")
plt.show()
````

````python
import matplotlib.pyplot as plt

# Assuming the dataframe is named df
df['income'] = df['income'].astype(float)  # Ensure the income column is float

# Sort the dataframe by state and then by income
df = df.sort_values(['name', 'income'])

# Calculate the year-over-year income growth rate for each state
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Calculate the average income growth rate for each state
avg_income_growth = df.groupby('name')['income_growth'].mean()

# Convert the series to a dataframe
avg_income_growth_df = avg_income_growth.reset_index()

# Plot the average income growth rate for each state
plt.figure(figsize=(10,8))
plt.bar(avg_income_growth_df['name'], avg_income_growth_df['income_growth'])
plt.xlabel('State')
plt.ylabel('Average Income Growth Rate')
plt.title('Average Year-Over-Year Income Growth Rate by State')
plt.xticks(rotation=90)
plt.show()
````
